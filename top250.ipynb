```python
import pandas as pd
import numpy as np
import json
import numpy as np
from imdb import IMDb
```

```python
movies = pd.read_csv("ml-latest-small/linked_movies.csv")
```

```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
movie_genre = pd.DataFrame()
movie_language = pd.DataFrame()
```

```python
for i in range(0,len(movies)):
    movie =  movies.loc[i]
    movie_id = movie['movieId']
    genres = movie['genres'].split('|')
    df = pd.DataFrame({movie_id:np.ones(len(genres))}, index = genres)
    movie_genre = pd.concat([movie_genre, df], axis=1, sort=False)
```

```python
ia = IMDb()
```

```python
top250 = ia.get_top250_movies()
```

```python
top250_movies = {}
for movie in top250:
    top250_movies[int(movie.getID())] = movie.data['rating']
```

```python
top250_movies = pd.DataFrame.from_dict(top250_movies, orient='index', columns=['imdb_rating'])
```

```python
top250_movies_ratings = top250_movies['imdb_rating'].values
```

```python
links = pd.read_csv("ml-latest-small/links.csv")
```

```python
top250_movies
```

```python
from top250_imdb import Content
model = Content(ratings, movie_genre, top250_movies, links, movies)
```

```python
recs = model.recommendations
```

```python
with open('recommendations/top250_imdb/top250_imdb.json', 'w') as json_file:
    json.dump(recs, json_file)
```
